```python
#question 2 

"A scatter plot is a type of data visualization that displays individual data points as dots on a two-dimensional Cartesian plane. Each dot represents a single observation in a dataset and is positioned based on the values of two variables – one on the x-axis and the other on the y-axis."

import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x,y)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Scatter Plot")
```

```python
#question 3
import matplotlib.pyplot as plt
import numpy as np

data = [
    (np.array([0, 1, 2, 3, 4, 5]), np.array([0, 100, 200, 300, 400, 500]), "Line 1"),
    (np.array([0, 1, 2, 3, 4, 5]), np.array([50, 20, 40, 20, 60, 70]), "Line 2"),
    (np.array([0, 1, 2, 3, 4, 5]), np.array([10, 20, 30, 40, 50, 60]), "Line 3"),
    (np.array([0, 1, 2, 3, 4, 5]), np.array([200, 350, 250, 550, 450, 150]), "Line 4")
]

plt.figure(figsize=(10, 8))

for i, (x, y, label) in enumerate(data, 1):
    plt.subplot(2, 2, i)
    plt.plot(x, y, marker='o')
    plt.title(label)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.show()
```

```python
#question 4 

"A bar plot, also known as a bar chart, is a type of data visualization that uses rectangular bars to represent data values. Each bar corresponds to a category or group, and the length or height of the bar is proportional to the value it represents. Bar plots are particularly effective for comparing the values of different categories or groups."

import numpy as np
import matplotlib.pyplot as plt
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company,profit)
```

```python
#question 5 

"A box plot, also known as a box-and-whisker plot, is a type of data visualization that displays the distribution and statistical summary of a dataset. It provides a concise way to visualize the central tendency, spread, and potential outliers within a dataset. Box plots are particularly useful for identifying the shape of the distribution and highlighting any unusual data points."

import numpy as np 
import matplotlib.pyplot as plt
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data=[box1,box2]
plt.boxplot(data)
```

```python
#question 1 

import numpy as np 
import scipy.stats as stats

sample_mean = 50
sample_std = 5 
sample_size = 100

standard_error = sample_std /np.sqrt(sample_size)

zscore = stats.norm.ppf(0.975)

MOE = zscore * standard_error

lower_bound = sample_mean - MOE
upper_bound = sample_mean + MOE

print(upper_bound,lower_bound)
```

```python
#question 4 

import scipy.stats as stats

sample_size = 500
sample_proportion = 60 / sample_size  


standard_error = (sample_proportion * (1 - sample_proportion) / sample_size)**0.5


z_score = stats.norm.ppf(0.975) 
margin_of_error = z_score * standard_error
lower_bound = sample_proportion - margin_of_error
upper_bound = sample_proportion + margin_of_error
lower_bound_percentage = lower_bound * 100
upper_bound_percentage = upper_bound * 100
print(f"95% Confidence Interval for the True Proportion of Smokers: {lower_bound_percentage:.2f}% to {upper_bound_percentage:.2f}%")

```

```python
#question 6 

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats


df = 10
x = np.linspace(0, 30, 1000)
pdf = stats.chi2.pdf(x, df)
plt.plot(x, pdf, label=f"Chi-Square (df={df})")
x_shaded = np.linspace(0, 15, 1000)
plt.fill_between(x_shaded, 0, stats.chi2.pdf(x_shaded, df), where=(x_shaded <= 15), color='red', alpha=0.5, label="Chi-Square Statistic of 15")


plt.xlabel("Chi-Square Value")
plt.ylabel("Probability Density")
plt.legend()
plt.show()
```

```python
#question 8 

observed = 45
total_flips = 100
expected = total_flips * 0.5
df=1
alpha = 0.05
chi_sqaure_statistics = (observed - expected)**2/expected 

p_value = 1 - stats.chi2.cdf(chi_sqaure_statistics,df) 

if p_value < alpha:
    print("Reject the null hypothesis. The coin appears to be biased towards tails.")
else:
    print("Fail to reject the null hypothesis. There is no strong evidence that the coin is biased towards tails.")
```

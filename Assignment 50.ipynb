```python
def removal_box_plot(df, Glucose, threshold):
    sns.boxplot(df[Glucose])
    plt.title(f'Original Box Plot of {Glucose}')
    plt.show()
 
    removed_outliers = df[df[Glucose] <= threshold]
 
    sns.boxplot(removed_outliers[Glucose])
    plt.title(f'Box Plot without Outliers of {Glucose}')
    plt.show()
    return removed_outliers 
 
threshold_value = 250
removal_box_plot(df, 'Glucose', threshold_value)
```

```python
#question 3 

X = df.iloc[::-1]
y = df['Outcome']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

```

```python
#question 4 

from sklearn.tree import DecisionTreeClassifier
treeclassifier=DecisionTreeClassifier()
treeclassifier.fit(X_train,y_train)
```

```python
#question 5 

from sklearn.metrics import confusion_matrix,classification_report

clf = treeclassifier.fit(X_train, y_train)
predictions = clf.predict(X_test)
cm = confusion_matrix(y_test, predictions, labels=clf.classes_)
cm
```

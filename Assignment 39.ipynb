```python
#question 13 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")

%matplotlib inline

df=pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')
df.set_index
df.head()

chars_to_remove=['+',',','$']
cols_to_clean=['Installs','Price']
for item in chars_to_remove:
    for cols in cols_to_clean:
        df[cols]=df[cols].str.replace(item,'')

df = df[df['Price'] != 'Everyone']

df['Price'] = df['Price'].astype('float')

average_prices = df.groupby('Type')['Price'].mean()
plt.figure(figsize=(8, 6))
average_prices.plot(kind='bar', color=['green', 'red'])  
plt.show()
```

```python
#question 14 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")

%matplotlib inline

df=pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')
df.set_index
df.head()

chars_to_remove=['+',',','$']
cols_to_clean=['Installs','Price']
for item in chars_to_remove:
    for cols in cols_to_clean:
        df[cols]=df[cols].str.replace(item,'')
df = df.sort_values(by='Installs', ascending=False)
top_10_apps = df.head(10)
print(top_10_apps[['App', 'Installs']])
df.head()
```

```python
#question 17 

df = df[df['Price'] != 'Everyone']

df['Last Updated']=pd.to_datetime(df['Last Updated'])
df['Day']=df['Last Updated'].dt.day
df['Month']=df['Last Updated'].dt.month
df['Year']=df['Last Updated'].dt.year

launch_counts_by_month = df.groupby('Month')['App'].count()
launch_counts_by_month.plot(kind='line', marker='o')
plt.title('Number of App Launches by Month')
plt.xlabel('Month')
plt.ylabel('Number of App Launches')
plt.show()
```
